```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample DataFrame for context
data = {'Sex': ['male', 'female', 'male', 'female', 'male', 'female', 'female']}
df = pd.DataFrame(data)

# Define colors (hex color and named color)
colors = ['#C4061D', 'green']

# Create the pie chart
plt.figure(figsize=(8, 6))
plt.pie(df['Sex'].value_counts(), labels=['Male', 'Female'], colors=colors, shadow=True, autopct='%.2f%%', startangle=140)

# Remove axis
plt.axis('off')

# Add a legend with a title
plt.legend(title='Sex', loc='best')

# Display the pie chart
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample DataFrame for context
data = {
    'Age': [22, 38, 26, 35, 35, 28, 2, 14, 4, 58],
    'Survived': [0, 1, 1, 1, 0, 1, 0, 1, 1, 0],
    'Sex': ['male', 'female', 'female', 'female', 'male', 'female', 'male', 'female', 'male', 'male']
}
df = pd.DataFrame(data)

# Set the figure size
plt.figure(figsize=(10, 6))

# Create the barplot
sns.barplot(x='Age', y='Survived', hue='Sex', data=df, palette='Set2')

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Survival Rate')
plt.title('Survival Rate by Age and Sex')

# Rotate x-axis labels for better readability if needed (e.g., if there are many age groups)
plt.xticks(rotation=45)

# Display the plot
plt.show()
```
